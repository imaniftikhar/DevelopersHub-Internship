```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Marvel Vs DC.csv')
```

```python
df
```

```python
df=df['Description']
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
```

```python
vectors=vectorizer.fit_transform(df)
```

```python
vectors
```

```python
bow_df=pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names_out() )
```

```python
bow_df
```

```python
word_count=bow_df.sum(axis=0)
```

```python
word_count
```

```python
word_count_df=pd.DataFrame(word_count, columns=['counts']).sort_values(by='counts',ascending=False)
```

```python
word_count_df=word_count_df.head(20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.bar(word_count_df.index,word_count_df['counts'] )
plt.xlabel('words')
plt.ylabel('Total Count')
plt.title('Total Countingf Bar Graph')
plt.xticks(rotation=45)
plt.show()
```
