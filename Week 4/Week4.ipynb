# Scraped Data Visualization

```python
import pandas as pd
```

```python
df=pd.read_csv('Output.csv')
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
top_df=df.head(15)
```

```python
top_df['Revenue growth'] = top_df['Revenue growth'].str.rstrip('%').astype(float)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Revenue (USD millions)', y='Name', data=top_df, palette='viridis')
plt.title('Top U.S. Companies by Revenue', fontsize=16)
plt.xlabel('Revenue (USD millions)', fontsize=12)
plt.ylabel('Company Names', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(top_df['Revenue growth'],kde=False)
plt.show()
```

```python
sns.distplot(top_df['Revenue growth'], hist=False)
plt.show()
```

```python
if (top_df['Revenue growth'] < 0).any():
    print("Warning: Negative values found in revenue data! These will be excluded from the pie chart.")
    top_df = top_df[top_df['Revenue growth'] >= 0]  # Exclude negative values

# Now create the pie chart
plt.figure(figsize=(8,8))
plt.pie(top_df['Revenue growth'], labels=top_df['Name'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Revenue Distribution Among Top Companies')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=top_df['Revenue (USD millions)'], color='lightblue')
plt.title('Revenue Box Plot for Top U.S. Companies', fontsize=16)
plt.ylabel('Revenue (USD millions)', fontsize=12)
plt.show()
```
